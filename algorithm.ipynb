```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from pathlib import Path
from io import StringIO


if Path('only_resoults.csv').exists():
    pass
else:
    # read the data
    train_data = pd.read_csv('train.csv')
    test_data = pd.read_csv('test.csv')
    
    categorys = ['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked']
    
    # using one-hot encoding to make all the data numeric
    encoder = OneHotEncoder(handle_unknown='ignore')
    encoder.fit(train_data[categorys])
    y_train = train_data['Survived']
    X_train = train_data.drop(columns=['Survived'])
    X_train = encoder.transform(X_train[categorys])
    testing_data = encoder.transform(test_data[categorys])
    
    # repleacing the 'NaN' values
    imputer = SimpleImputer(strategy='mean')
    
    testing_data = imputer.fit_transform(testing_data)
    X_train = imputer.fit_transform(X_train)
    
    # feading the final data and making predictions
    dtc = DecisionTreeClassifier()
    dtc.fit(X_train, y_train)
    y_pred = dtc.predict(testing_data)
    
    # create the resoults file
    passengerid = test_data['PassengerId'].to_list()
    passengerid = np.array(passengerid)
    survived_column = pd.DataFrame({'Survived': y_pred})
    passengerid_column = pd.DataFrame({'PassengerId': passengerid})
    resoults = survived_column.join(passengerid_column)
    
    # convert the data to a csv
    csv_file = resoults.to_csv(index=False)
    string_io = StringIO(csv_file)
    csv_bytes = string_io.getvalue().encode()
    
    Path('only_resoults.csv').write_bytes(csv_bytes)
```
